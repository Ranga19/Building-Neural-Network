```python
import pandas as pd
from sklearn.model_selection import train_test_split 
import tensorflow as tf
from tensorflow import keras
import math
import numpy as np
```

```python
data = pd.read_csv("insurance_data.csv")
```

```python
print(data.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[['age', 'affordibility']], data.bought_insurance, test_size = 0.2, random_state = 25)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age'] / 100
x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age'] / 100
```

```python
#building model with tensorflow and then we will build the gradient descent model manually.

model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation = 'sigmoid', kernel_initializer = 'ones', bias_initializer = 'zeros')
])

model.compile(
    optimizer = "adam",
    loss = "binary_crossentropy",
    metrics = ['accuracy']
)
model.fit(x_train_scaled, y_train, epochs = 4500)
```

```python
model.evaluate(x_test_scaled, y_test)
```

```python
print("x_test")
print(x_test_scaled)
print("y_test")
print(y_test)

model.predict(x_test_scaled)
```

```python
coef, intercept = model.weights
```

```python
print(coef, intercept)

# we got w1=4.59, w2=1.35 and b=-2.67
```

```python
#building model manually

'''activation function---sigmoid
 y = w1*x1 + w2*x2 + b
 z = 1 / (1 + (e^-x))'''

def sigmoid(x):
    return 1 / (1 + np.exp(-x))



'''log loss or binary crossentropy for calcluting total loss
    loss i = -(y log(y^)) + (1-y)log(1-y^)'''

def logloss(y_pred, y_true):
    return -np.mean(y_true * np.log(y_pred)) + (1 - y_true) * np.log(1 - y_pred)


def gradient_descent(x_train_scaled, y_train, epochs):
    w1, w2, b = 1, 1, 0
    age = x_train_scaled['age']
    affor = x_train_scaled['affordibility']
    learning_rate = 0.05
    loss = 0
    for i in range(epochs):
        sum = w1 * age + w2 * affor + b
        y_pred = sigmoid(sum)
        n = len(age)
        dldw1 = 1 / n * np.dot(np.transpose(age), (y_pred - y_train))
        dldw2 = 1 / n * np.dot(np.transpose(affor), (y_pred - y_train))
        dldb = np.mean(y_pred - y_train)
        
        w1 = w1 - learning_rate * dldw1
        w2 = w2 - learning_rate * dldw2
        b = b - learning_rate * dldb
        loss = float(logloss(y_pred, y_train)[0])
        print("epoch:{0}, w1:{1}, w2:{2}, b:{3}, loss:{4}".format(i, w1, w2, b, loss))
    return (w1, w2, b)
```

```python
w1, w2, b = gradient_descent(x_train_scaled, y_train, 1000)
print(w1, w2, b)
```

```python
#testing with manual input

age = float(input("Enter age:")) / 100
affor = int(input("Enter 0 or 1 (affordibility):"))
y_pred = age * w1 + affor * w2 + b
print("sigmoid function value:",y_pred)
if y_pred > 0.5:
    res = 1
else:
    res = 0
print("result:", res)
```
